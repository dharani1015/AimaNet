```python
#Video duration in seconds
import cv2

def video_seconds(path):
    
    video_data = cv2.VideoCapture(path)
    video_frames = video_data.get(cv2.CAP_PROP_FRAME_COUNT)
    video_fps = video_data.get(cv2.CAP_PROP_FPS)
    vid_seconds = round(video_frames/video_fps)
    
    return vid_seconds
```

```python
import numpy as np
from skimage.util import img_as_float
import cv2
from tqdm import tqdm
import matplotlib.pyplot as plt
from numpy import moveaxis

def preprocess_vid(vdp, dim = 36, show = False):
    
    if show:
        print("Video Path:", vdp)
        check_video(vdp)
                  

    videoFilePath = vdp


    if show:
        print("File Exists:", os.path.isfile(videoFilePath))
    #########################################################################
    t = []
    i = 0
    vidObj = cv2.VideoCapture(videoFilePath);
    totalFrames = int(vidObj.get(cv2.CAP_PROP_FRAME_COUNT))
    height = vidObj.get(cv2.CAP_PROP_FRAME_HEIGHT)
    width = vidObj.get(cv2.CAP_PROP_FRAME_WIDTH)
    success, img = vidObj.read()
    dims = img.shape

    if show:
        print("Total Frames:", totalFrames)
        print("Orignal Height", height)
        print("Original width", width)
        print("Dims:", dims)

        print("#"*50)
        print("Read Image/Frame Properties:")
        print(type(img))
        print(img.shape)
        print("Max:", img.max())
        print("Min:", img.min())
        print("Mean:", img.mean())
        print("#"*50)

    #########################################################################
    # Crop each frame size into dim x dim

    Xsub = np.zeros((totalFrames, dim, dim, 3), dtype = np.float32)

    count = 0
    while success:
        t.append(vidObj.get(cv2.CAP_PROP_POS_MSEC))

        vidLxL = cv2.resize(img_as_float(img), (dim, dim), interpolation = cv2.INTER_AREA) #Norms the image.

        vidLxL = cv2.cvtColor(vidLxL.astype('float32'), cv2.COLOR_BGR2RGB)
        vidLxL[vidLxL > 1] = 1
        vidLxL[vidLxL < (1/255)] = 1/255
        Xsub[i, :, :, :] = vidLxL
        success, img = vidObj.read() # read the next one
        i = i + 1

    plt.imshow(Xsub[0])
    plt.title('Sample Preprocessed Frame')
    plt.show()
    
    if show:
        print("Xsub Shape:", Xsub.shape)
        print("Max:",Xsub.max())
        print("Min:",Xsub.min())

    #########################################################################
    # Normalized Frames in the motion branch
    normalized_len = Xsub.shape[0] - 1
    dXsub = np.zeros((normalized_len, dim, dim, 3), dtype = np.float32)
    for j in range(normalized_len - 1):
        dXsub[j, :, :, :] = (Xsub[j+1, :, :, :] - Xsub[j, :, :, :]) / (Xsub[j+1, :, :, :] + Xsub[j, :, :, :])

    if show:
        print("dXsub Shape:", dXsub.shape)
        print("Max:",dXsub.max())
        print("Min:",dXsub.min())
    
    #########################################################################
    Xsub = Xsub[:totalFrames-1, :, :, :]

    #########################################################################
    dXsub = np.concatenate((dXsub, Xsub), axis = 3);

    #Move Axis
    if show:
        print("Final Shape: (CL)", dXsub.shape)
    dXsub = moveaxis(dXsub, -1, 1)
    if show:
        print("Final Shape: (CF):", dXsub.shape)
        print("Max:",dXsub.max())
        print("Min:",dXsub.min())
        
    return dXsub
```

#Preprocessing with Face Detection

```python
import numpy as np
from skimage.util import img_as_float
import cv2
from tqdm import tqdm
import matplotlib.pyplot as plt
from numpy import moveaxis

import mediapipe as mp

mp_face = mp.solutions.face_detection.FaceDetection(
    model_selection=0, # model selection 5m range
    min_detection_confidence=0.5 # confidence threshold
)

def preprocess_vid(vdp, dim = 36, show = False):
    
    if show:
        print("Video Path:", vdp)
        check_video(vdp)
                  
    videoFilePath = vdp


    if show:
        print("File Exists:", os.path.isfile(videoFilePath))
    #########################################################################
    # set up
    t = []
    i = 0
    vidObj = cv2.VideoCapture(videoFilePath);
    totalFrames = int(vidObj.get(cv2.CAP_PROP_FRAME_COUNT)) # get total frame size
    height = vidObj.get(cv2.CAP_PROP_FRAME_HEIGHT)
    width = vidObj.get(cv2.CAP_PROP_FRAME_WIDTH)
    success, img = vidObj.read()
    dims = img.shape

    if show:
        print("Total Frames:", totalFrames)
        print("Orignal Height", height)
        print("Original width", width)
        print("Dims:", dims)

        print("#"*50)
        print("Read Image/Frame Properties:")
        print(type(img))
        print(img.shape)
        print("Max:", img.max())
        print("Min:", img.min())
        print("Mean:", img.mean())
        print("#"*50)

    #########################################################################
    # Crop each frame size into dim x dim

    Xsub = np.zeros((totalFrames, dim, dim, 3), dtype = np.float32)

    count = 0
    while success:
        t.append(vidObj.get(cv2.CAP_PROP_POS_MSEC))# current timestamp in milisecond
        try:
          img_face = img
            #############################################
            #Face Detection
          results = mp_face.process(img_face)
          height, width, channels = img.shape

          if results.detections:
            for detection in results.detections: # iterate over each detection and draw on image
              bbox = detection.location_data.relative_bounding_box
              xmin = int(bbox.xmin * width)
              ymin = int(bbox.ymin * height)
              xmax = int(bbox.width * width + bbox.xmin * width)
              ymax = int(bbox.height * height + bbox.ymin * height)
              xmargin = 20
              ymargin = 30
              img_face = img_face[ymin - ymargin:ymax + ymargin, xmin - xmargin:xmax + xmargin]

              break

        except Exception as e:
          img_face = img

        vidLxL = cv2.resize(img_as_float(img_face), (dim, dim), interpolation = cv2.INTER_AREA) #Norms the image.

        vidLxL = cv2.cvtColor(vidLxL.astype('float32'), cv2.COLOR_BGR2RGB)
        vidLxL[vidLxL > 1] = 1
        vidLxL[vidLxL < (1/255)] = 1/255
        Xsub[i, :, :, :] = vidLxL
        success, img = vidObj.read() # read the next one
        i = i + 1

    plt.imshow(Xsub[0])
    plt.title('Sample Preprocessed Frame')
    plt.show()
    
    if show:
        print("Xsub Shape:", Xsub.shape)
        print("Max:",Xsub.max())
        print("Min:",Xsub.min())

    #########################################################################
    # Normalized Frames in the motion branch
    normalized_len = Xsub.shape[0] - 1
    dXsub = np.zeros((normalized_len, dim, dim, 3), dtype = np.float32)
    for j in range(normalized_len - 1):
        dXsub[j, :, :, :] = (Xsub[j+1, :, :, :] - Xsub[j, :, :, :]) / (Xsub[j+1, :, :, :] + Xsub[j, :, :, :])

    if show:
        print("dXsub Shape:", dXsub.shape)
        print("Max:",dXsub.max())
        print("Min:",dXsub.min())
    
    #########################################################################
    
    Xsub = Xsub[:totalFrames-1, :, :, :]

    #########################################################################
    dXsub = np.concatenate((dXsub, Xsub), axis = 3);

    #Move Axis
    if show:
        print("Final Shape: (CL)", dXsub.shape)
    dXsub = moveaxis(dXsub, -1, 1)
    if show:
        print("Final Shape: (CF):", dXsub.shape)
        print("Max:",dXsub.max())
        print("Min:",dXsub.min())
        
    return dXsub
```

```python
def check_video(vp):
  print("Input:",vp)
  cap = cv2.VideoCapture(vp)
  frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
  print("Number of frames:", frames )
  fps = cap.get(cv2.CAP_PROP_FPS)
  print("FPS:",fps)
  ts = round(frames/fps, 2)
  tm = round((frames/fps) / 60, 2)
  print("Video Duration (s):", ts,'seconds')
  print("Video Duration (m):", tm,'mins')

  return frames, fps, ts
```
